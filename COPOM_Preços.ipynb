```python
import requests
import pandas as pd
from datetime import datetime


today = datetime.today().strftime('%Y-%m-%d')

first_meeting = '2024-09-19'
second_meeting = '2024-11-07'
third_meeting = '2024-12-12'

df = pd.DataFrame()
meetings  = [first_meeting, second_meeting, third_meeting]
for i in meetings:
    url = f'https://arquivos.b3.com.br/charts/chart/copom/{today}/{i}'
    r = requests.get(url)
    df = pd.concat([df, pd.DataFrame(r.json())])

df = df.reset_index(drop=True)
df
```
